```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
f = 5  # Hz
t_cont = np.linspace(0, 1, 1000)  # Continuous time
t_sample = np.linspace(0, 1, 50)  # Sampled points

# Signals
y_cont = np.sin(2 * np.pi * f * t_cont)
y_sample = np.sin(2 * np.pi * f * t_sample)

# Plot
plt.figure(figsize=(10,4))
plt.plot(t_cont, y_cont, label="Continuous Sine Wave")
plt.stem(t_sample, y_sample, linefmt='r-', markerfmt='ro', basefmt='k', label="Sampled Points")
plt.title("5 Hz Sine Wave - Continuous & Sampled")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.show()

# Display as 5x10 matrix
matrix_5x10 = y_sample.reshape(5,10)
print("Sampled Values (5x10 Matrix):\n", np.round(matrix_5x10,2))
```

```python
# Quantization
levels = 8
y_quantized = np.round((y_sample - y_sample.min()) / (y_sample.max()-y_sample.min()) * (levels-1)) * ((y_sample.max()-y_sample.min())/(levels-1)) + y_sample.min()

# Plot
plt.figure(figsize=(10,4))
plt.stem(t_sample, y_sample, linefmt='b-', markerfmt='bo', basefmt='k', label='Sampled')
plt.stem(t_sample, y_quantized, linefmt='r--', markerfmt='ro', basefmt='k', label='Quantized')
plt.title("Sampled vs Quantized Sine Wave")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.show()
```

```python
custom_img = np.array([[0, 50, 100, 150, 200, 250, 0, 50, 100, 150],[50, 100, 150, 200, 250, 0, 50, 100, 150, 200],
                       [100, 150, 200, 250, 0, 50, 100, 150, 200, 250],[150, 200, 250, 0, 50, 100, 150, 200, 250, 0],
                       [200, 250, 0, 50, 100, 150, 200, 250, 0, 50],[250, 0, 50, 100, 150, 200, 250, 0, 50, 100],
                       [0, 50, 100, 150, 200, 250, 0, 50, 100, 150],[50, 100, 150, 200, 250, 0, 50, 100, 150, 200],
                       [100, 150, 200, 250, 0, 50, 100, 150, 200, 250],[150, 200, 250, 0, 50, 100, 150, 200, 250, 0]])

plt.figure(figsize=(12,3))
plt.subplot(1,4,1)
plt.imshow(custom_img, cmap='gray', vmin=0, vmax=255)
plt.title("Original")

down1 = custom_img[::2, ::2]
down2 = custom_img[::3, ::3]
down3 = custom_img[::4, ::4]

plt.subplot(1,4,2)
plt.imshow(down1, cmap='gray', vmin=0, vmax=255)
plt.title("Downsample x2")

plt.subplot(1,4,3)
plt.imshow(down2, cmap='gray', vmin=0, vmax=255)
plt.title("Downsample x3")

plt.subplot(1,4,4)
plt.imshow(down3, cmap='gray', vmin=0, vmax=255)
plt.title("Downsample x4")
plt.show()

print("Downsampled x2:\n", down1)
print("Downsampled x3:\n", down2)
print("Downsampled x4:\n", down3)
```

```python
img_4x4 = np.array([[0, 64, 128, 192],
                    [32, 96, 160, 224],
                    [16, 80, 144, 208],
                    [48, 112, 176, 240]])

levels_list = [256, 16, 8, 4, 2]
plt.figure(figsize=(15,3))

for i, L in enumerate(levels_list):
    q_img = np.floor(img_4x4/(256/L))*(256/L)
    plt.subplot(1,5,i+1)
    plt.imshow(q_img, cmap='gray', vmin=0, vmax=255)
    plt.title(f'{L} Levels')
    print(f"{L}-Level Quantized Image:\n", q_img)

plt.show()
```
