```python
#sampling with matrix
#Plots the original continuous sine wave and the sampled points.

#Creates a matrix (5x10) from the sampled signal (since you have 50 samples).

#Rounds values to 2 decimal places for clarity.
import numpy as np
import matplotlib.pyplot as plt

# Continuous signal (sine wave)
t = np.linspace(0, 1, 1000)   # high resolution (like infinite fabric)
signal = np.sin(2 * np.pi * 5 * t)  # 5 Hz sine wave

# Sampling at fixed intervals
sample_rate = 50   # samples per second
t_sampled = np.linspace(0, 1, sample_rate)
signal_sampled = np.sin(2 * np.pi * 5 * t_sampled)

# Plot the continuous and sampled signal
plt.figure(figsize=(10, 5))
plt.plot(t, signal, label="Continuous Signal", color="blue")
plt.stem(t_sampled, signal_sampled, linefmt='r-', markerfmt='ro', basefmt=" ", label="Sampled Points")
plt.legend()
plt.title("Sampling a Continuous Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()

# Show sampled signal as matrix (rounded for readability)
matrix = np.round(signal_sampled.reshape(5, 10), 2)  # reshape into 5x10 matrix
print("Sampled Signal Matrix (5x10):")
print(matrix)
```

```python
#Quantization (Mapping values to finite levels)
# Quantization function
def quantize(signal, levels):
    min_val, max_val = signal.min(), signal.max()
    q_levels = np.linspace(min_val, max_val, levels)
    quantized = np.digitize(signal, q_levels) - 1
    return q_levels[quantized]

# Quantize the sampled signal
levels = 8   # e.g., 3-bit quantization
quantized_signal = quantize(signal_sampled, levels)

# Plot
plt.figure(figsize=(10, 5))
plt.stem(t_sampled, signal_sampled, linefmt='g-', markerfmt='go', basefmt=" ", label="Sampled Signal")
plt.stem(t_sampled, quantized_signal, linefmt='r-', markerfmt='ro', basefmt=" ", label="Quantized Signal")
plt.legend()
plt.title(f"Quantization of Sampled Signal ({levels} Levels)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
# @title
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a simple synthetic image
img = np.zeros((60, 160), dtype=np.uint8)  # black background
cv2.rectangle(img, (20, 20), (50, 50), 255, 2)

# --- Arithmetic Operations ---
brightened = cv2.add(img, 100)   # Increase brightness
darkened   = cv2.subtract(img, 100)  # Decrease brightness

# --- Logical Operations ---
# Create a circular mask
mask = np.zeros_like(img, dtype=np.uint8)
cv2.circle(mask, (100, 100), 60, 255, -1)

and_result = cv2.bitwise_and(img, mask)
or_result  = cv2.bitwise_or(img, mask)
xor_result = cv2.bitwise_xor(img, mask)
not_result = cv2.bitwise_not(img)

# --- Display results ---
fig, axes = plt.subplots(3, 3, figsize=(10, 10))
axes = axes.ravel()

titles = ["Original", "Brightened (+100)", "Darkened (-100)",
          "Mask (Circle)", "AND with Mask", "OR with Mask",
          "XOR with Mask", "NOT (Inverted)", ""]

images = [img, brightened, darkened,
          mask, and_result, or_result,
          xor_result, not_result, None]

for i, ax in enumerate(axes):
    if images[i] is not None:
        ax.imshow(images[i], cmap="gray")
    ax.set_title(titles[i])
    ax.axis("off")

plt.tight_layout()
plt.show()
```

```python
#Simulate Video as Sequence of Frames
import numpy as np
import matplotlib.pyplot as plt

# Create 5 frames with moving square
frames = []
for t in range(5):
    img = np.zeros((20,20))
    img[5+t:10+t, 5:10] = 255
    frames.append(img)

# Display frames
fig, axes = plt.subplots(1,5, figsize=(12,3))
for i, ax in enumerate(axes):
    ax.imshow(frames[i], cmap="gray")
    ax.set_title(f"Frame {i+1}")
    ax.axis("off")
plt.show()
```

```python
#Compare Audio Formats (WAV vs MP3)
from scipy.io.wavfile import write
from pydub import AudioSegment

# Generate audio
fs = 44100
t = np.linspace(0,1,fs)
x = np.sin(2*np.pi*440*t)
write("sound.wav", fs, (x*32767).astype(np.int16))

# Convert to MP3
sound = AudioSegment.from_wav("sound.wav")
sound.export("sound.mp3", format="mp3")

import os
print("WAV size:", os.path.getsize("sound.wav"), "bytes")
print("MP3 size:", os.path.getsize("sound.mp3"), "bytes")
```

```python
#Color Models (RGB  HSV  YUV)
import cv2
import matplotlib.pyplot as plt

# Create red RGB patch
rgb = np.zeros((100,100,3), dtype=np.uint8)
rgb[:,:] = [255,0,0]  # Red

hsv = cv2.cvtColor(rgb, cv2.COLOR_RGB2HSV)
yuv = cv2.cvtColor(rgb, cv2.COLOR_RGB2YUV)

fig, axs = plt.subplots(1,3, figsize=(9,3))
axs[0].imshow(rgb); axs[0].set_title("RGB"); axs[0].axis("off")
axs[1].imshow(hsv); axs[1].set_title("HSV"); axs[1].axis("off")
axs[2].imshow(yuv); axs[2].set_title("YUV"); axs[2].axis("off")
plt.show()
```

```python
# Demonstrate Audio Properties (Sampling Rate, Frequency, Amplitude)
import numpy as np
import matplotlib.pyplot as plt

# Same tone, different sampling rates
f = 440
dur = 0.01
fs1, fs2 = 2000, 8000

t1 = np.linspace(0, dur, int(fs1*dur))
t2 = np.linspace(0, dur, int(fs2*dur))

x1 = np.sin(2*np.pi*f*t1)
x2 = np.sin(2*np.pi*f*t2)

plt.figure(figsize=(8,4))
plt.plot(t1, x1, "o-", label="fs=2000 Hz")
plt.plot(t2, x2, ".", label="fs=8000 Hz")
plt.legend()
plt.title("Effect of Sampling Rate on Audio Signal")
plt.show()
```
